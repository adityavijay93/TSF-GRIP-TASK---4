### Data Visualization

```python
# Histograms to check Killed and Wounded
plt.figure(figsize=(45,20))
df_terrorism.hist(figsize=(20,10))
plt.tight_layout()
```

```python
# Visualizing the correlation between the Attributes
plt.figure(figsize=(15,10))
sns.heatmap(df_terrorism.corr(), annot=True)
```

```python
# Visualizing the Number Of Terrorist Activities Each Year
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=df_terrorism,palette='YlOrRd',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
```

```python
# Visualization to check the Attacking Methods by Terrorists
plt.subplots(figsize=(15,6))
sns.countplot('AttackType',data=df_terrorism,palette='inferno',order=df_terrorism['AttackType'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Attacking Methods by Terrorists')
```

```python
# Visualiztion to see Targets of Terrorists
plt.subplots(figsize=(15,6))
sns.countplot(df_terrorism['Target_type'],palette='inferno',order=df_terrorism['Target_type'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Targets of Terrorists')
```

```python
# Number Of Terrorist Attacks Per Region 
plt.subplots(figsize=(15,6))
sns.countplot('Region',data=df_terrorism,palette='RdYlGn',edgecolor=sns.color_palette('dark',7),order=df_terrorism['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Attacks Per Region')
```

```python
# Number Of Terrorist Attacks Per Country 
plt.subplots(figsize=(15,6))
sns.countplot('Country',data=df_terrorism,palette='RdYlGn',edgecolor=sns.color_palette('dark',7),order=df_terrorism['Country'].value_counts()[:30].index)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Attacks Per Country')
plt.tight_layout()
```

```python
# Counts of Attacks vs Killed
count_terror=df_terrorism['Country'].value_counts()[:30].to_frame()
count_terror.columns=['Attacks']
count_kill=df_terrorism.groupby('Country')['Killed'].sum().to_frame()
count_terror.merge(count_kill,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(18,6)
```

```python
# Terrorist Groups with Highest Terror Attacks
plt.subplots(figsize=(11,10))
sns.barplot(y=df_terrorism['Group'].value_counts()[1:30].index,x=df_terrorism['Group'].value_counts()[1:30].values,
           palette='inferno')
plt.title('Terrorist Groups with Highest Terror Attacks')
```
